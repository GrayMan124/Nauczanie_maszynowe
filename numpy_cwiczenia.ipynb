```python
def almost_variance(x):
    """Oblicza 1/n * SUM (x_i - mean(x))^4"""
    m = sum(x) / len(x)
    result = 0
    for i in range(len(x)):
        result += (x[i] - m) ** 4
    result /= len(x)
    return result

def numpy_almost_var(x):
    m = np.mean(x)
    powers = (x - m) ** 4
    return np.mean(powers)

X = np.random.random(10000)
```

```python
%%timeit
almost_variance(X)
```

```python
%%timeit
numpy_almost_var(X)
```
